## fitting the data to Polynomial regressor

#### adding necessary imports for polynomial regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
monthly_polynomial_regressor = []
polyregList = []
X_List = []
y_List=[]
```

```python
for i in range(0,12):
    X = Monthly_Crime_Rate_Data[i].iloc[:15, 1:2].values
    y = Monthly_Crime_Rate_Data[i].iloc[:15, 0].values
    X_List.append(X)
    y_List.append(y)
    
    poly_reg = PolynomialFeatures(degree = 3)
    X_poly = poly_reg.fit_transform(X)
    poly_reg.fit(X_poly, y)
    lin_reg_2 = LinearRegression()
    lin_reg_2.fit(X_poly, y)
    monthly_polynomial_regressor.append(lin_reg_2)
    polyregList.append(poly_reg)
```

```python
len(monthly_polynomial_regressor)
```

```python
type(monthly_polynomial_regressor[0])
```

### Validation on old data & prediction on new Data

```python
monthName = ["January" ,"Febuary" ,"March" , "April" , "May" , "June" , "July" , "August" ,"September" ,"October"
            ,"November","December"]
predictedYear = [2016 ,2017 ,2018 ,2019 ,2020 ,2025]
```

```python
k=0
y_true_polynomial = []
y_pred_polynomial = []
for i in range(0 ,12):
    print("\t\t"+monthName[i])
    print("Predicted"+"\t\t\t"+"Validation Data")
    result = monthly_polynomial_regressor[i].predict(poly_reg.fit_transform([[2016],[2017],[2018],[2019],[2020],[2025]]))
    for j in range(0,6):
        if(k<2):
            y_pred_polynomial.append(result[j])
            y_true_polynomial.append(Monthly_Crime_Rate_Data[i][Monthly_Crime_Rate_Data[i].Year==predictedYear[j]].CrimeRate.values[0])
            k=k+1
        if(j<3):
            print(str(predictedYear[j])+" -> "+str(result[j])+"\t\t"+str(Monthly_Crime_Rate_Data[i][Monthly_Crime_Rate_Data[i].Year==predictedYear[j]].CrimeRate.values[0]))
        else :
            print(str(predictedYear[j])+" -> "+str(result[j]))
    k=0
    print("\n\n")
```

### Visulising the regression curve

```python
for i in range(0,12):
    X_grid = np.arange(min(X_List[i]), max(X_List[i]), 0.1)
    X_grid = X_grid.reshape((len(X_grid), 1))
    plt.scatter(X_List[i], y_List[i], color = 'red')
    plt.plot(X_grid, monthly_polynomial_regressor[i].predict(polyregList[i].fit_transform(X_grid)), color = 'blue')
    plt.title(monthName[i]+" "+'Trends')
    plt.xlabel('Year')
    plt.ylabel('Crime Rates')
    plt.savefig(monthName[i]+"_polynomial_Regression")
    plt.show()
```

## fitting the data to Support Vector Regressor

#### adding necessary imports for svr 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
```

```python
monthly_svr_regressor = []
X_List = []
y_List=[]
sc_X_List = []
sc_Y_List = []
X_Scaled_List = []
y_Scaled_List=[]
```

```python
for i in range(0,12):
    X = Monthly_Crime_Rate_Data[i].iloc[:15, 1:2].values
    y = Monthly_Crime_Rate_Data[i].iloc[:15, 0].values
    X_List.append(X)
    y_List.append(y)
    
    sc_X = StandardScaler()
    sc_y = StandardScaler()
    X = sc_X.fit_transform(X)
    y = sc_y.fit_transform(y)
    sc_X_List.append(sc_X)
    sc_Y_List.append(sc_y)
    X_Scaled_List.append(X)
    y_Scaled_List.append(y)
    
    regressor = SVR(kernel = 'rbf')
    regressor.fit(X, y)
    monthly_svr_regressor.append(regressor)
```

```python
len(monthly_svr_regressor)
```

```python
type(monthly_svr_regressor[0])
```

### Validation on old data & prediction on new Data

```python
k=0
y_true_SVR=[]
y_pred_SVR =[]
for i in range(0 ,12):
    print("\t\t"+monthName[i])
    print("Predicted"+"\t\t\t"+"Validation Data")
    y_pred = monthly_svr_regressor[i].predict(sc_X_List[i].transform(np.array([[2016],[2017],[2018],[2019],[2020],[2025]])))
    result = sc_Y_List[i].inverse_transform(y_pred)
    for j in range(0,6):
        if(k<2):
            y_pred_SVR.append(result[j])
            y_true_SVR.append(Monthly_Crime_Rate_Data[i][Monthly_Crime_Rate_Data[i].Year==predictedYear[j]].CrimeRate.values[0])
            k=k+1
        if(j<3):
            print(str(predictedYear[j])+" -> "+str(result[j])+"\t\t"+str(Monthly_Crime_Rate_Data[i][Monthly_Crime_Rate_Data[i].Year==predictedYear[j]].CrimeRate.values[0]))
        else :
            print(str(predictedYear[j])+" -> "+str(result[j]))
    k=0
    print("\n\n")
```

### Visualising the svr curve

```python
for i in range(0,12):
    X_grid = np.arange(min(X_Scaled_List[i]), max(X_Scaled_List[i]), 0.01)
    X_grid = X_grid.reshape((len(X_grid), 1))
    plt.scatter(X_Scaled_List[i], y_Scaled_List[i], color = 'red')
    plt.plot(X_grid,monthly_svr_regressor[i].predict(X_grid), color = 'blue')
    #plt.plot(X_grid, sc_Y_List[i].inverse_transform(monthly_svr_regressor[i].predict(sc_X_List[i].fit_transform(X_grid))), color = 'blue')
    plt.title(monthName[i]+" "+'Trends')
    plt.xlabel('Year')
    plt.ylabel('Crime Rates')
    plt.savefig(monthName[i]+"_svr")
    plt.show()
```

## Model Evaluation 

### Polynomial Regression & Support Vector Regression

#### Explained Variance Score 

```python
#The explained_variance_score computes the explained variance regression score.
#If y hat(^) is the estimated target output, y the corresponding (correct) target output, and Var is Variance, 
#the square of the standard deviation, then the explained variance is estimated as follow:

#{explained\_{}variance}(y, \hat{y}) = 1 - \frac{Var\{ y - \hat{y}\}}{Var\{y\}}

#The best possible score is 1.0, lower values are worse.
```

```python
from sklearn.metrics import explained_variance_score
print("Explained Variance Score on SVR Model : "+str(explained_variance_score(y_true_SVR, y_pred_SVR) ))
print("Explained Variance Score on Polynomial Regression Model : "+str(explained_variance_score(y_true_polynomial, y_pred_polynomial) ))
```

#### Mean Absolute Error

```python
#The mean_absolute_error function computes mean absolute error, a risk metric corresponding to the expected value of the absolute error loss 

#If yi hat(^) is the predicted value of the i-th sample, and  is the corresponding true value, 
#then the mean absolute error (MAE) estimated over n_sample is defined as

# {MSE}(y, \hat{y}) = \frac{1}{n_\text{samples}} \sum_{i=0}^{n_\text{samples} - 1} (y_i - \hat{y}_i)^2.
```

```python
from sklearn.metrics import mean_absolute_error
print("Mean absolute error on SVR Model : "+str(mean_absolute_error(y_true_SVR, y_pred_SVR) ))
print("Mean absolute error on Polynomial Regression Model : "
      +str(mean_absolute_error(y_true_polynomial, y_pred_polynomial)))
```

### R² score, the coefficient of determination

```python
#The r2_score function computes R², the coefficient of determination. 
#It provides a measure of how well future samples are likely to be predicted by the model. 
#Best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse). 
#A constant model that always predicts the expected value of y, disregarding the input features, 
#would get a R^2 score of 0.0.
```

```python
from sklearn.metrics import r2_score
print("R² score on SVR Model : "+str(r2_score(y_true_SVR, y_pred_SVR) ))
print("R² score on Polynomial Regression Model : "+str(r2_score(y_true_polynomial, y_pred_polynomial)))
```
