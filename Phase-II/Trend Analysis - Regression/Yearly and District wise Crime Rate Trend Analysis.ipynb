## fitting the data to Support Vector Regressor

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
```

```python
monthly_svr_regressor = []
X_List = []
y_List=[]
sc_X_List = []
sc_Y_List = []
X_Scaled_List = []
y_Scaled_List=[]
```

## Model Evaluation 

### Polynomial Regression & Support Vector Regression

#### Explained Variance Score 

```python
#The explained_variance_score computes the explained variance regression score.
#If y hat(^) is the estimated target output, y the corresponding (correct) target output, and Var is Variance, 
#the square of the standard deviation, then the explained variance is estimated as follow:

#{explained\_{}variance}(y, \hat{y}) = 1 - \frac{Var\{ y - \hat{y}\}}{Var\{y\}}

#The best possible score is 1.0, lower values are worse.
```

```python
from sklearn.metrics import explained_variance_score
print("Explained Variance Score on SVR Model : "+str(explained_variance_score(y_true_svr, y_pred_svr) ))
print("Explained Variance Score on Polynomial Regression Model : "+str(explained_variance_score(y_true_polynomial, y_pred_polynomial) ))
```

#### Mean Absolute Error

```python
#The mean_absolute_error function computes mean absolute error, a risk metric corresponding to the expected value of the absolute error loss 

#If yi hat(^) is the predicted value of the i-th sample, and  is the corresponding true value, 
#then the mean absolute error (MAE) estimated over n_sample is defined as

# {MSE}(y, \hat{y}) = \frac{1}{n_\text{samples}} \sum_{i=0}^{n_\text{samples} - 1} (y_i - \hat{y}_i)^2.
```

```python
from sklearn.metrics import mean_absolute_error
print("Mean absolute error on SVR Model : "+str(mean_absolute_error(y_true_svr, y_pred_svr) ))
print("Mean absolute error on Polynomial Regression Model : "
      +str(mean_absolute_error(y_true_polynomial, y_pred_polynomial)))
```

### R² score, the coefficient of determination

```python
#The r2_score function computes R², the coefficient of determination. 
#It provides a measure of how well future samples are likely to be predicted by the model. 
#Best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse). 
#A constant model that always predicts the expected value of y, disregarding the input features, 
#would get a R^2 score of 0.0.
```

```python
from sklearn.metrics import r2_score
print("R² score on SVR Model : "+str(r2_score(y_true_svr, y_pred_svr) ))
print("R² score on Polynomial Regression Model : "+str(r2_score(y_true_polynomial, y_pred_polynomial)))
```
